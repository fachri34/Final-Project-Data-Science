### A-1 Dropping missing values berpersentase kecil

```python
print(f'Jumlah baris sebelum mendrop missing values: {len(df)}')
df.dropna(inplace=True, subset=['enrolled_university','education_level','experience','last_new_job'])
print(f'Jumlah baris setelah mendrop missing values: {len(df)}')
```

### A-2 Imputasi data di kolom dengan presentase missing values besar

```python
df[['major_discipline','company_size','company_type']].mode()
```

```python
# Gender --> others
# major_discipline --> modus: STEM
# company_size = 50-99
# company_type = Pvt Ltd

df['gender'] = df['gender'].fillna('others')
df['major_discipline'] = df['major_discipline'].fillna('STEM')
df['company_size'] = df['company_size'].fillna('50-99')
df['company_type'] = df['company_type'].fillna('Pvt Ltd')

df.isna().sum()
```

### A-3 Mengubah Kolom2 dengan data yang belum sesuai

```python
df.sample(5)
```

```python
#1 mengubah enrollee_id menjadi string
df['enrollee_id'] = df['enrollee_id'].astype(str)

#2
df['city']= df['city'].str.replace('city_','')
df['city']= df['city'].astype(int)

#3
df['relevent_experience'] = df['relevent_experience'].apply(lambda x: '1' if x=='Has relevent experience' else '0')

#4
df['experience'] = df['experience'].apply(lambda x: '0' if x=='<1' else x)
df['experience'] = df['experience'].apply(lambda x: '21' if x=='>20' else x)
df['experience'] = df['experience'].astype(int)


#5
#company_size': akan dibersihkan dan dikelompokkan menjadi 8 kelompok (<10, 10-49, 50-99, 100-500, 500-999, 1000-4999, 5000-999, 10000+)
df['company_size'] = df['company_size'].apply(lambda x: '10-49' if x=='10/49' else x)

#6
df['last_new_job'] = df['last_new_job'].apply(lambda x: '5' if x=='>4' else x)
df['last_new_job'] = df['last_new_job'].apply(lambda x: '0' if x=='never' else x)
df['last_new_job'] = df['last_new_job'].astype(int)

#7
df['target'] = df['target'].astype(int)
```

```python
df.sample(10)
```

### C. Handling outliers

```python
# Cek distribusi kolom numerik
cats = ['enrollee_id','city', 'gender', 'relevent_experience', 'enrolled_university', 'education_level','major_discipline','company_size', 
        'company_type','last_new_job']
nums = ['city_development_index','experience','training_hours','target']

print(df[cats].describe())
print("\n", df[nums].describe())
```

```python
plt.figure(figsize=(8,3))
for i, cols in enumerate(nums,1):
    plt.subplot(1,4,i)
    sns.boxplot(data=df[nums], y=cols, orient='v',color='Green')
plt.tight_layout()
```

```python
from scipy import stats
```

```python
print(f'Jumlah baris sebelum memfilter outlier: {len(df)}')

filtered_entries = np.array([True] * len(df))

for col in ['city_development_index','experience','training_hours']:
    zscore = abs(stats.zscore(df[col])) # hitung absolute z-scorenya
    filtered_entries = (zscore < 3) & filtered_entries # keep yang kurang dari 3 absolute z-scorenya
    
df = df[filtered_entries] # filter, cuma ambil yang z-scorenya dibawah 3

print(f'Jumlah baris setelah memfilter outlier: {len(df)}')
```

```python
plt.figure(figsize=(12,3))
for i,cols in enumerate(nums,1):
    plt.subplot(1,4,i)
    sns.kdeplot(df[cols])
    plt.tight_layout()
```

### E. Feature Encoding

```python
df.head()
```

```python
print(df['enrolled_university'].unique())
print(df['education_level'].unique())
print(df['company_size'].unique())
```

```python
# Kolom2 yang harus diencode:
# label encoding:
# gender, relevant experience, enrolled_university, education_level, company_size 
# one hots:
# major_discipline, company_type
```

```python
# label encoder pada gender, relevant experience, enrolled university, education level, company size.
df['gender'] = df['gender'].astype('category').cat.codes
df['relevent_experience'] = df['relevent_experience'].astype('category').cat.codes

urut_eu= ['no_enrollment', 'Part time course', 'Full time course']
urut_el= ['Primary School', 'High School', 'Graduate', 'Masters', 'Phd']
urut_cs= ['<10','10-49', '50-99','100-500','500-999','1000-4999','5000-9999','10000+']

label_mapping_eu = {kategori: label for label, kategori in enumerate(urut_eu)}
df['label_eu'] = df['enrolled_university'].map(label_mapping_eu)
label_mapping_el = {kategori: label for label, kategori in enumerate(urut_el)}
df['label_el'] = df['education_level'].map(label_mapping_el)
label_mapping_cs = {kategori: label for label, kategori in enumerate(urut_cs)}
df['label_cs'] = df['company_size'].map(label_mapping_cs)

df.head()
```

```python
# major_discipline, company_type
for cat in ['major_discipline', 'company_type']:
  onehots = pd.get_dummies(df[cat], prefix=cat)
  df = df.join(onehots)
```

```python
df_train = pd.read_csv('aug_train.csv')

sns.set(style='whitegrid')

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
plt.subplots_adjust(wspace=0.4, hspace=0.4)

sns.histplot(ax=axes[0, 0], x=df_train['city_development_index'], bins=30, kde=True, color='skyblue')
axes[0, 0].set_title('Distribution of City Development Index')
axes[0, 0].set_xlabel('City Development Index')
axes[0, 0].set_ylabel('Frequency')
sns.boxplot(ax=axes[1, 0], x=df_train['city_development_index'], color='lightgrey', width=0.2)

sns.histplot(ax=axes[0, 1], x=df_train['training_hours'], bins=30, kde=True, color='skyblue')
axes[0, 1].set_title('Distribution of Training Hours')
axes[0, 1].set_xlabel('Training Hours')
axes[0, 1].set_ylabel('Frequency')
sns.boxplot(ax=axes[1, 1], x=df_train['training_hours'], color='lightgrey', width=0.2)

sns.histplot(ax=axes[0, 2], x=df_train['target'], bins=2, kde=True, color='skyblue')
axes[0, 2].set_title('Distribution of Target')
axes[0, 2].set_xlabel('Target')
axes[0, 2].set_ylabel('Frequency')
sns.boxplot(ax=axes[1, 2], x=df_train['target'], color='lightgrey', width=0.2)

Q1 = df_train['city_development_index'].quantile(0.25)
Q3 = df_train['city_development_index'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df_train[(df_train['city_development_index'] < lower_bound) | (df_train['city_development_index'] > upper_bound)]

sns.boxplot(ax=axes[0, 3], x=df_train['city_development_index'], orient='vertical', color='lightgrey')
axes[0, 3].scatter(x=outliers.index, y=outliers['city_development_index'], color='skyblue', label='Outliers')
axes[0, 3].set_title('Boxplot of City Development Index with Outliers')
axes[0, 3].set_xlabel('City Development Index')
axes[0, 3].set_ylabel('')
axes[0, 3].legend()

plt.tight_layout()

plt.show()
```
